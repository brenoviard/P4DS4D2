```python
# %Load https://matplotlib.or/_downloads/pyplot_text.py

"""
===========
Pyplot Text
===========

"""

import numpy as np
import matplotlib.pyplot as plt


np.random.seed(19680801)


mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n,bis,pacthes = plt.hist(x, 50, density=True, facecolor='b', alpha=0.45)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
