## Stings em dataframes


Usualmente não costuma-se trabalhar com dados tipos string apenas.
É mais comum que usemos dataframes, que são conjuntos de dados.

```python
import pandas as pd

df = pd.read_csv('../titanic.csv')
```

```python
df.head()
```

```python
df_exemplo = pd.DataFrame([('C115235', 100, 25, 'paulo@gmail.com'),
('B215235', 200, 34, 'maria@gmail.com'),
('a215235', 380, 40, 'fatima@gmail.com'),
('B825235', 150, 25, 'pedro@hotmail.com'),
('a825235', 250, 30, 'joao@yahoo.com'),
('A8252355', 400, 45, 'carolina@yahoo.com')],
columns=('Id', 'Vlr_Gasto', 'Idade', 'Email'))
df_exemplo
```

```python
df_exemplo['Id'].str
```

```python
df_exemplo['Qtd_carcateres_id'] = df_exemplo['Id'].str.len()
df_exemplo
```

```python
df_exemplo['Id'] = df_exemplo['Id'].str.upper()
df_exemplo
```

```python
for n in range(0, len(df_exemplo)):
    df_exemplo.loc[n,'DDD'] = df_exemplo.loc[n,'Id'][1:3]
for n in range(0, len(df_exemplo)):
    df_exemplo.loc[n,'Risco'] = df_exemplo.loc[n,'Id'][0]
    
df_exemplo
```

```python
for  n in range(0,len(df_exemplo)):
    df_exemplo.loc[n,'cod'] = df_exemplo.loc[n,'DDD'] + df_exemplo.loc[n,'Risco']
df_exemplo
```

```python
df_exemplo['email_ajust'] = df_exemplo['Email'].str.split('@').str[0]
df_exemplo
```

Removedo o primeiro caraceter de id

```python
df_exemplo['Id_Ajuste'] = df_exemplo['Id'].str.slice(1)
df_exemplo
```

Removendo o último caractere de id

```python
df_exemplo['Id_Novo'] = df_exemplo['Id'].str.slice(start=0,stop=7)
df_exemplo
```

Concatenando e criando um novo id
 -- em caso dos valores serem numericos e string, é necessário converter numericos em string
 

```python
df_exemplo['Id_Final'] = df_exemplo['Id_Novo']+df_exemplo['Email']
df_exemplo
```

Localizar/identificar substrings 

```python
df_exemplo['FL_RJ'] = df_exemplo['Id'].str.find('21')
df_exemplo
```

Encontrando quem tem Gmail

```python
df_exemplo['provedor_gmail'] = df_exemplo['Email'].str.find('gmail')
df_exemplo
```
