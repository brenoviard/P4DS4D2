Removendo 'lbs' da coluna peso
Necessário a importação da bibilioteca re
O que foi feito:
    criado uma nova variável onde será recebido os valores de weight sem 'lbs
    a função map lê os valores com base na função lambda criada
    e a partir disso ela substitui 'lbs' por ''

```python
import re
df['Weight_New'] = df['Weight'].map(lambda x: re.sub(r'lbs','',x))
df.head(2)
```

Substituindo NaN

```python
df['Weight_New'] = df['Weight_New'].map(lambda x: re.sub(r'nan','9999',x))
```

 Convertendo os valores da coluna de string para numeric

```python
df['Weight_New']=pd.to_numeric(df.Weight_New)
df['Weight_New'].dtypes
```

### Tratando a clausula de rescisão

Cria uma nova coluna com os valores da Release_Clause em Milhares

```python
# Verifica se K é maiúsculo ou minusculo

df[df['Release Clause'].str.contains("K", na = False)]
```

Converte o Camplo Release Clause em string

```python
df['Release Clause'] = df['Release Clause'].astype('str')
```

Se for valor em M, coluna milhoes recebeo valor e a coluna mil não

```python
df.loc[df['Release Clause'].str.contains('M'), 'Valor_M'] = df['Release Clause'].map(lambda x: re.sub(r'M','',x))
df.loc[~df['Release Clause'].str.contains('M'), 'Valor_M'] = 0
```

```python
df.loc[df['Release Clause'].str.contains('K'), 'Valor_K'] = df['Release Clause'].map(lambda x: re.sub(r'K','',x))
df.loc[~df['Release Clause'].str.contains('K'), 'Valor_K'] = 0
                                                                                     
```

```python
df['Valor_K'].dtype
```

Converte os valores das novas colunas para string

```python
df['Valor_K'] = df['Valor_K'].astype('str')
df['Valor_M'] = df['Valor_M'].astype('str')
```

Removendo o simbolo de euros

```python
df['Valor_K'] = df['Valor_K'].map(lambda x: re.sub(r'€','',x))
df['Valor_M'] = df['Valor_M'].map(lambda x: re.sub(r'€','',x))
```

Converter as colunas para numero

```python
df['Valor_K'] = pd.to_numeric(df.Valor_K)
df['Valor_M'] = pd.to_numeric(df.Valor_M)
```

Somando Valor_M * 1000 + Valor_K

```python
df['Valor_Final(K)'] = df['Valor_M'] * 1000 + df['Valor_K']
df.tail()
```
