Você possui as informações de clientes que foram até a loja A e realizaram o cadastro em algum momento, também possui informações de clientes que se cadastraram na loja B e possui uma terceira base com todas informações de compras já feitas em qualquer uma das duas lojas:

```python
# Cadastro da loja a
cadastro_a = {'Id': ['AA2930','BB4563','CC2139','DE2521','GT3462','HH1158'],
'Nome': ['Andre', 'Amanda', 'Bruna', 'Carlos', 'Jose', 'Maria'],
'Idade': [20,35,40,54,29,25],
'CEP': ['00092-029','11111-111','22222-888','00000-999','99999-111','55555-666']
}

cadastro_a = pd.DataFrame(cadastro_a, columns = ['Id','Nome','Idade','CEP'])
print(cadastro_a)

# Cadastro da loja b
cadastro_b = {'Id': ['CC9999','EF4488','DD9999','GT3462','HH1158'],
'Nome': ['Marcos', 'Patricia', 'Rafaela', 'Jose', 'Maria'],
'Idade': [19,30,22,29,25],
'CEP': ['00092-029','11111-111','22222-888','99999-111','55555-666']
}

cadastro_b = pd.DataFrame(cadastro_b, columns = ['Id','Nome','Idade','CEP'])
print(cadastro_b)

# Registro de compras de todas unidades
compras = {'Id': ['AA2930','EF4488','CC2139','EF4488','CC9999','AA2930','HH1158','HH1158'],
'Data': ['2019-01-01','2019-01-30','2019-01-30','2019-02-01','2019-02-20','2019-03-15','2019-04-01','2019-04-10'],
'Valor': [200,100,40,150,300,25,50,500]
}

compras = pd.DataFrame(compras, columns = ['Id','Data','Valor'])
print(compras)

```

É fundamental que o nome das colunas no dataframe sejam iguais ao do dicionário com os dados

### Sintaxe do Merge
Merge é uma biblioteca do Pandas
 - é necessário fornecer as bases 
 - indicar a coluna que será utilizada para unir(qual coluna as tabelas possuem em comum)
 - e o tipo de join a ser realizado
 
 #### pd.merge(tabela_equerda, tabela_direita, on='coluna_em_comum', how='left|right|inner|outer')
 
 Havendo a necessidade de usar mais de uma coluna, fazemos:
#### pd.merge(tabela_da_esquerda, tabela_da_direita, on=["coluna_1","coluna_2"], how="left|right|inner|outer")

 Caso os nomes das colunas em comum serem diferentes, usamonos left_on e right_on
#### pd.merge(tabela_da_esquerda, tabela_da_direita, right_on=["coluna_direita_1","coluna_direita_2"], left_on=["coluna_esquerda_1","coluna_esquerda_2"], how="left|right|inner|outer")


### Inner Join + Alteração nos nomes das colunas

```python
pd.merge(cadastro_a,cadastro_b,
        on = 'Id',
        how = 'inner')
```

Dessa forma ele traz todas as informações de ambas as tabelas

```python
pd.merge(cadastro_a, cadastro_b[['Id','CEP','Idade']],
        on = ['Id'],
        how = 'inner')
```

Aqui selecionamos da tabela B apenas cep e idade dos clientes

```python
pd.merge(cadastro_a,
        cadastro_b[['Id','Idade','CEP']],
        on = 'Id',
        how = 'inner',
        suffixes = ('_A','_B'))
```

Para aleterar os sufixos dos nomes da colunas usamos o parametro suffixes.
Agora, caso queiramos alterar toda a tabelas, fazemos:


```python
inner_join = pd.merge(cadastro_a,
                     cadastro_b[['Id','Idade','CEP']],
                     on = 'Id',
                     how = 'inner')
inner_join.columns = ['Id','Nome','Idade','CEP','Idade_B','CEP_B']
inner_join
```

```python
pd.merge(cadastro_a, compras,
        on = 'Id',
        how = 'left')
```

### Full Join

Full join é diferente do Outer Join
Para unir as duas tabelas de cadastro, utilizamos o full join.
Apesar de semelhante, o outer não atendeiria a nossa necessidade.
Nesse caso, usaremos o concat:

```python
pd.concat([cadastro_a,cadastro_b], ignore_index = True)
```

Perceba que jose e maria foram duplicados, para resolver use o drop_duplicates:

```python
empilhados = pd.concat([cadastro_a,cadastro_b], ignore_index = True)

empilhados.drop_duplicates(subset='Id')
```

### Left Join

Para identificar quais clientes compraram e o valor gasto

```python
left = pd.merge(cadastro_a, compras, on = 'Id',
        how = 'left')
left = left[left['Valor'].notna()]

left = left.groupby(['Id','Nome']).agg(total = ('Valor', sum))
left
```

Agora que sabemos quais clientes fizeram compras, podemos utilizar o groupby para ver o total gasto por esses clientes.
